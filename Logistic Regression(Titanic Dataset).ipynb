```python
highly_correlated = []


for cols in df1.columns:
    print(cols)
    x=df1[f"{cols}"].corr(df1['Survived']) #f string will actually replace values of each column
    print(round(x,2))
    if abs(x) > 0.2:
        highly_correlated.append(f"{cols}")
   
```

```python
X = df1.drop('Survived',axis=1)   # drop all independent cols from df1 in X axis
y = df1[['Survived']]             # select dependent col in y axis
```

# 7. Divide components into X and Y and Train Test Split

```python
# Train Test Split
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=100)
```

### 8. Scaling


```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.fit_transform(X_test)
```

### 9. Training the Data (with the appropriate alogorithm from the right family alogrithms)

```python
from sklearn.linear_model import LogisticRegression 
# importing the right algorithm from the right model
lg = LogisticRegression()
```

```python
#fitting the data
lg.fit(X_train,y_train)
```

### 10. Testing the Data with y_test

```python
#testing
y_pred = lg.predict(X_test)
```

```python
y_pred
```

### 11. Evaluation of the final model with right metrics ( R2, RMSE, MAE, MSE , adj. R2)

```python
from sklearn.metrics import ConfusionMatrixDisplay, classification_report
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```
